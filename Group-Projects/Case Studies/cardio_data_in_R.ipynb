## Cardio Analysis

This is an analysis of 5 months of step data from my phone. I input the data into a Google doc and saved it as a CSV file so it could be uploaded to R Studio. There I cleaned it and made the visualizations below.

First I separated the date column into month, day, year so the month could be isolated. Then I found the average steps for each month.

```r
library(tidyverse)
```

```r
library(ggplot2)
```

```r
cardio <- read_csv("https://raw.githubusercontent.com/CSIS-NLU/Open-Science-101/main/Group-Projects/Case%20Studies/steps.csv")
```

```r
head(cardio)
```

```r
cardio %>%
  separate(Date, sep="/", into = c("month", "day", "year"))
```

```r
head(cardio)
```

```r
cardio %>%
    summarise(average_steps=mean(Steps), min_steps=min(Steps), max_steps=max(Steps))
```

## Visualizations of Step Data
I have tried everything to get this ggplot code to work but for some reason my visualizations are not viewable. If anyone has some suggestions please let me know.

Below is a scatter plot illustrating daily steps per month for 5 months in 2021.

```r
ggplot(data = cardio, aes(x = Steps, y = Date, color = Steps)) +
  geom_point()
```

```r
options(repr.plot.width =20, repr.plot.height =20)
ggplot(data = cardio, mapping = aes(x = Steps, y = Date, color=Steps))+
geom_point()+
theme(axis.text.x = element_text(angle = 90)) +
    theme(plot.title = element_text(face="bold"))+
    labs(title="Number of Steps by Day", subtitle = "Steps from 8/21-12/21", x="Date", y="Steps per Day")
```
