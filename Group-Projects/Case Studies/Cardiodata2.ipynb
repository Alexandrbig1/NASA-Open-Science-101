Histograms

```python
cardiodata['Date'] = pd.to_datetime(cardiodata.Date)
```

```python
cardiodata.hist(figsize=(20,20))
```

Boxplots

```python
sns.boxenplot(x='Date',y='Steps',data=cardiodata)
```

Average Steps

```python
cardiodata = cardiodata.replace(',','', regex=True)
cardiodata.head()
```

```python
cardiodata.info()
```

```python
cardiodata['Steps'] = cardiodata['Steps'].astype(float)
```

```python
cardiodata['Steps'].mean()
```

```python
sns.distplot(cardiodata['Steps'])
```

```python
sns.pairplot(cardiodata)
```

```python
sns.scatterplot(x='Steps', y='Date',data=cardiodata)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
sns.scatterplot(x='Date', y='Steps',data=cardiodata, s=50, ax=ax)
plt.title('Scatterplot showing steps by date')
plt.show()
```
